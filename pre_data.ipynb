```python
import pandas as pd
import numpy as np
```

# 划分数据集 并保存

```python
data1_df = pd.read_csv('data1.csv',encoding='utf8')
data1_df
```

```python
train_index_df= pd.read_csv('train_index.csv')
train_index=np.array(train_index_df['train'])
test_index_df= pd.read_csv('test_index.csv')
test_index=np.array(test_index_df['test'])
```

```python
train_index.shape
```

```python
test_index.shape
```

```python
y_data = data1_df['是否早发']
y_data
```

```python
image_id = data1_df['image_ID']
```

```python
x_data = data1_df.drop('是否早发',axis = 1)
y_data_copy = y_data
```

```python
from imblearn.over_sampling import SMOTE, ADASYN
data1_df, y_data = SMOTE().fit_sample(data1_df, y_data_copy)
x_data, y_label = SMOTE().fit_sample(x_data, y_data_copy)
```

```python
data1_df['位置'] = data1_df['位置'].astype(str)

pd.get_dummies(data1_df['位置'], prefix='位置').head()
```

```python
data1_df['强化方式'] = data1_df['强化方式'].astype(str)

pd.get_dummies(data1_df['强化方式'], prefix='强化方式').head()
```

```python

data1_df['乙肝or丙肝'] = data1_df['乙肝or丙肝'].astype(str)

pd.get_dummies(data1_df['乙肝or丙肝'], prefix='乙肝or丙肝').head()
```

```python
data1_df_new = pd.get_dummies(data1_df)
data1_df_new.head()
```

```python
data1_df_new.shape
```

```python
from sklearn.model_selection import train_test_split
#x_train, x_test, y_train,y_test=train_test_split(data1_df,y_data,test_size=0.2,random_state=42,stratify=y_data)
x_train, x_test, y_train,y_test = data1_df_new.iloc[train_index],data1_df_new.iloc[test_index],y_data.iloc[train_index],y_data.iloc[test_index]

```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train_image_id = x_train['image_ID']
x_train_image_id
```

```python
x_test_image_id = x_test['image_ID']
x_test_image_id
```

```python
x_train_label = x_train['是否早发']
x_train_label
```

```python
x_test_label = x_test['是否早发']
x_test_label
```

## 临床

```python
c_values1 = x_train.iloc[:,0:21]
c_values2 = x_train.iloc[:,1688+1688+21:]
clinical_traindata = df = pd.concat( [c_values1,c_values2], axis=1 )
clinical_traindata
```

```python
c_values1 = x_test.iloc[:,0:21]
c_values2 = x_test.iloc[:,1688+1688+21:]
clinical_testdata = df = pd.concat( [c_values1,c_values2], axis=1 )
clinical_testdata
```

```python
clinical_traindata.to_csv('clinical_train.csv',index=False,header=True)
clinical_testdata.to_csv('clinical_test.csv',index=False,header=True)
```
